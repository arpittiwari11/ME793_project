# COEFF_A

```python
model_scores = { 'GRP' : 0.9737245457009842,
'RIDGE' :  0.9500177370824315,
'RANDOM_FOREST' : 0.9438097954730255,
'XGBoost' : 0.914104096236886,
'NN' : 0.9525409428164175 }
```

```python

import matplotlib.pyplot as plt

# Extract model names and scores
models = list(model_scores.keys())
r_squared_scores = list(model_scores.values())

# Plot the bar graph
plt.figure(figsize=(10, 6))
plt.bar(models, r_squared_scores, color='orange')
plt.xlabel('Models')
plt.ylabel('R-squared Score')
plt.title('Comparison of R-squared Scores')
plt.ylim(0, 1)  # Set y-axis limit from 0 to 1 (assuming R-squared scores range from 0 to 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
